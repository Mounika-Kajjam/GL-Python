```python
# Load the dataset
#read_csv --> to load th csv files on python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/titanic.csv")
```

```python
data.head()
```

```python
data = data.sort_values(by = 'Age', ascending=False)
```

```python
# Drops the records where the data is missing

#Drop the missing values, store the data into a new variable
df = data.dropna() # data = data.dropna()
df.shape
```

```python
df.isnull().sum()
```

```python
data.isnull().sum()
```

```python
# fillna --> Fill the missing values with a secified value

data.fillna(0, inplace=True)
data.isnull().sum()
```

```python
# drop --> Drops the Passenger column from the dataset
# axis=1 indicates drop with respect to column
data.drop('PassengerId', axis = 1, inplace = True)
data.head()
```

```python
data.shape
```

```python
# Drop the column name ticket and Parch

data.drop(['Parch', 'Ticket'], axis=1, inplace=True)
data.head()
```

```python
# Drop the rows with indices 630, 851 and 493

data.drop([630, 851, 493],axis=0, inplace=True)
data.head()
```
